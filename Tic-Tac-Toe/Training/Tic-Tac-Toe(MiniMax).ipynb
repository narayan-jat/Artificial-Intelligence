```python
# Creating tree
tree = {}
```

```python
def learning(state, player):
    lst = []            #List to store children of the state.  
    for i in range(9):
        if state[i] == 'N':
            next_state = create_state(state, player, i + 1)
            lst.append(next_state)
            status = is_terminal(next_state)
            if player == 'X':       #giving initial value to the states.
                if state not in tree:
                    tree[state] = 0
            if player == 'O':
                if state not in tree:
                    tree[state] = 0
            if status[0]:
                tree[next_state] = status[1]
                if player == 'X':       #Finding max value from children.
                    max = tree[state]
                    for s in lst:
                        if tree[s] > max:
                            max = tree[s]
                    tree[state] = max
                if player == 'O':       #Finding min value from children.
                    min = tree[state]
                    for s in lst:
                        if tree[s] < min:
                            min = tree[s]
                    tree[state] = min
            else:
                learning(next_state, 'O' if player == 'X' else 'X')     # Recursive call.
```

```python
# Learning the values of states.
learning('NNNNNNNNN', 'O')
learning('NNNNNNNNN', 'X')
```
